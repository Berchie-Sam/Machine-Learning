Categorical Data => qualitative data that can be grouped into categories instead of measured numerically and distict from each other.

```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/cars.csv', delimiter=';')
```

```python
df = dataset.head(10)
df
```

```python
df = dataset.tail(10)
df
```

Origin is categorical

```python
type(df)
```

```python
# Converting the dataframe to numpy array
df = df.iloc[:,:].values
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
country = LabelEncoder()
```

```python
df[:,8] = country.fit_transform(df[:,8])
```

```python
df
```

```python
# One hot encoding
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers=[('encode', OneHotEncoder(), [8])], remainder = 'passthrough')
```

```python
y = ct.fit_transform(df)
```

```python
y=pd.DataFrame(y)
y
```
