```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
url = '/content/drive/MyDrive/WA_Fn-UseC_-Telco-Customer-Churn.csv'
```

```python
data = pd.read_csv(url)
data
```

```python
data.columns
```

Data preparation

```python
data.columns = data.columns.str.lower().str.replace(' ', '_')

string_columns = list(data.dtypes[data.dtypes == 'object'].index)

for col in string_columns:
  data[col] = data[col].str.lower().str.replace(' ', '_')
```

```python
data
```

```python
pd.to_numeric(data.totalcharges, errors = 'coerce').fillna(0)
```

```python
from sklearn.model_selection import train_test_split
```

```python
data_train, data_test = train_test_split(data, test_size=0.2, random_state=1)
```

```python
data_train
```

```python
data_test
```

```python
data_train, data_val = train_test_split(data_train, test_size=0.25, random_state=1)
```

```python
len(data_train), len(data_val), len(data_test)
```

```python
# x = training data, y = target variable
y_train = (data_train.churn =='yes').astype('int')
y_val = (data_val.churn =='yes').astype('int')
y_test = (data_test.churn =='yes').astype('int')
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
del data_train['customerid']
del data_train['churn']
```

```python
del data_val['customerid']
del data_val['churn']
```

```python
dict_train = data_train.to_dict(orient='records')
dict_train[:5]
```

```python
dv = DictVectorizer()
dv.fit(dict_train)
```

```python
X_train = dv.transform(dict_train)
```

```python
feature_names = dv.get_feature_names_out()
print(feature_names)
```

```python
X_train[:5].todense()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train, y_train)
```

```python
dict_val = data_val.to_dict(orient='records')
dict_val[:5]
```

```python
X_val = dv.transform(dict_val)
```

```python
y_pred = model.predict(X_val)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_val, y_pred)
```

```python
import pickle
```

```python
with open('model.bin', 'wb') as f_out:
  pickle.dump((dv, model), f_out)
```
