```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Heart.csv')
df
```

```python
# Data preparation
df = df.drop(columns = 'Unnamed: 0')
df
```

```python
df['ChestPain'] = df['ChestPain'].astype('category')
df['ChestPain'] = df['ChestPain'].cat.codes
df
```

```python
df['Thal'] = df['Thal'].astype('category')
df['Thal'] = df['Thal'].cat.codes

df['AHD'] = df['AHD'].astype('category')
df['AHD'] = df['AHD'].cat.codes

df
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df
```

```python
x = df.drop(columns = 'AHD')
x
```

```python
y = df['AHD']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 21)
```

```python
print('x_train', x_train.shape)
print('x_test', x_test.shape)
print('y_train', y_train.shape)
print('y_test', y_test.shape)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
```

```python
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_scaled
```

```python
x_test_scaled
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression(random_state = 0).fit(x_train_scaled, y_train)
```

```python
log_reg.predict(x_train_scaled)
```

```python
log_reg.score(x_train_scaled, y_train)
```

```python
log_reg.score(x_test_scaled, y_test)
```

```python
log_reg1 = LogisticRegression(random_state = 0,
                              C = 2,
                              fit_intercept = True,
                              ).fit(x_train_scaled, y_train)
```

```python
train_score = log_reg1.score(x_train_scaled, y_train)
print(f'Training Score: {train_score:.4f}')
```

```python
test_score = log_reg1.score(x_test_scaled, y_test)
print(f'Testing Score: {test_score:.4f}')
```
