```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [10, 20, 30, 40, 50, 60, 70]

print(x)
print(y)

plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(1, 25, 10)
y = np.linspace(10, 1000, 10)

print(x)
print(y)
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)

# x and y labels, and title
plt.xlabel("x numbers")
plt.ylabel("y numbers")
plt.title("my 1st plot")
plt.show()
```

```python
# replacing line with some other symbol and introducing colors y is for yellow
plt.plot(x, y, '*')
plt.show()

plt.plot(x, y, 'y*')
plt.show()
```

```python
plt.plot(x, y, 'b', x, y**2, 'y>', x, y**3, '*')
plt.show()
```

```python
# Create a subplot with 1 row, 2 columns

# create the first subplot in figure 1
plt.subplot(131) # '1'=rows_number, '3'=columns_number, '1'='1st_subplot_position'
plt.title("1st subplot")
plt.plot(x, y)

# create the second subplot
plt.subplot(132) # 2-assigns the 2nd position to subplot 
plt.title("2nd subplot")
plt.plot(x, y**2)

# create the third subplot
plt.subplot(133) # 3-assigns the 2nd position to subplot
plt.title("3rd subplot")
plt.plot(x, y**3)
plt.show()
```

```python
# create a subplot with 2 rows and 2 columns
plt.figure(1)

plt.subplot(221)
plt.title('First')
plt.plot(x, y)

plt.subplot(222)
plt.title('Second')
plt.plot(x, y**2)

plt.figure(2)

plt.subplot(223)
plt.title('Third')
plt.plot(x, y**3)

plt.subplot(224)
plt.title('Fourth')
plt.plot(x, y**4)

plt.show()
```

##### Plotting categorical values and grouping values for meaningful insights and statistics

```python
# mergng market_fact and prod_dimen dataframes
mark_prod = pd.merge(market_fact, prod_dimen, how='inner', on='Prod_id')
mark_prod.head()
```

```python
# boxplot by product category
sns.boxplot(x='Product_Category', y='Sales', data=mark_prod)
plt.yscale('log')
plt.show()
```

```python
# boxplot by product category
sns.boxplot(x='Sales', y='Product_Category', data=mark_prod)
#plt.yscale('log')
plt.show()
```

```python
# boxplot by product category
sns.boxplot(x='Product_Category', y='Profit', data=mark_prod)
#plt.yscale('log')
plt.show()
```

```python
# Filtering out extremes

mark_prod1 = mark_prod[(mark_prod.Profit<800) & (mark_prod.Profit>-800)]
```

```python
# boxplot of a variable across various product categories
sns.boxplot(x='Product_Category', y='Profit', data=mark_prod1)
plt.show()
```

```python
## Customer segment plotting
# merging with customers df
mark_prod_cust = pd.merge(mark_prod, cust_dimen, how='inner', on='Cust_id')
mark_prod_cust.head()
```

```python
# boxplot of a variable across various product categories
sns.boxplot(x='Customer_Segment', y='Sales', data=mark_prod_cust)
plt.yscale('log')
plt.show()
```

```python
# showing outliers
sns.boxplot(x='Customer_Segment', y='Profit', data=mark_prod_cust)
plt.show()
```

```python
## filter out extremes
mark_prod_cust1 = mark_prod_cust[(mark_prod_cust.Profit<800) & (mark_prod_cust.Profit>-800)]
```

```python
sns.boxplot(x='Customer_Segment', y='Profit', data=mark_prod_cust1)
plt.show()
```

```python
# Visualizing across 2 caetgorical variables using hue argument
sns.boxplot(x='Customer_Segment', y='Profit', hue="Product_Category", data=mark_prod_cust1)
plt.show()
```

```python
plt.figure(num=None, figsize=(21,8))
```

```python
## Since the above image is very small, set the figure size to fit the image
plt.figure(num=None, figsize=(12, 8))
sns.boxplot(x='Customer_Segment', y='Profit', hue="Product_Category", data=mark_prod_cust1)
```

#### Plotting the aggregated values and grouping the results

```python
# bar plot
sns.barplot(x='Product_Category', y='Sales', data=mark_prod_cust1)
plt.title("Mean Sales")
plt.show()

# by default the aggregation bar plot plots the mean value
```

```python
# bar plot with default sta
sns.barplot(x='Product_Category', y='Sales', data=mark_prod_cust1, estimator=np.median)
plt.title("Median Plot")
plt.show()
```

```python
# using hue to add another category
plt.figure(num=None, figsize=(12, 8))

sns.barplot(x='Product_Category', y='Sales', hue='Customer_Segment', data=mark_prod_cust1, estimator=np.mean)
plt.show()

# small businesses are buying more technologies
# consumers are buying more furniture
```

```python
# plotting based on the sum of sales
plt.figure(num=None, figsize=(12, 8))

sns.barplot(x='Product_Category', y='Sales', hue="Customer_Segment", data=mark_prod_cust1, estimator=np.sum)
plt.show()
```

```python
# when plotting sub category, the x-axis names are not readable
plt#.figure(figsize(10, 8))
sns.barplot(x='Product_Sub_Category', y='Sales', data=mark_prod_cust1, estimator=np.sum)
plt.show()
```

```python
# reversing the x and y axes
plt.figure(figsize=(10, 8))
sns.barplot(y='Product_Sub_Category', x='Sales', data=mark_prod_cust1, estimator=np.sum)
plt.savefig('narplot1.png')
plt.show()
```
